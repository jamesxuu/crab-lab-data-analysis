```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the data from the Excel file
df = pd.read_excel('plotSpeedData.xlsx')

# Calculate the average and standard deviation of 'Speed (cm/cycle)' for each 'WaveNum'
grouped_df = df.groupby('WaveNum')['Speed (cm/cycle)']
avg_speed = grouped_df.mean()
std_dev = grouped_df.std()

# Plot the line chart with error bars using the default Matplotlib style
plt.errorbar(avg_speed.index, avg_speed, yerr=std_dev, fmt='o-', label='Flat Surface')
plt.xlabel('Wave Numbers')
plt.ylabel('Distance Traveled per Gait Cycle (cm/cycle)')
# plt.title('Flat Surface with Varied Wave Number')
plt.grid(True)

# Set x-axis ticks using the 'WaveNum' values from the DataFrame
plt.xticks(df['WaveNum'].unique())

# Show the legend with the label for the error bar
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Set the font family to 'sans-serif'
plt.rcParams['font.family'] = 'sans-serif'

# Read the data from the Excel file
df = pd.read_excel('plotSpeedDataRug.xlsx')

# Calculate the average and standard deviation of 'Speed (cm/cycle)' for each 'WaveNum'
grouped_df = df.groupby('WaveNum')['Speed (cm/cycle)']
avg_speed = grouped_df.mean()
std_dev = grouped_df.std()

# Plot the line chart with error bars using the default Matplotlib style
plt.errorbar(avg_speed.index, avg_speed, yerr=std_dev, fmt='o-', label='White Rugose', color='orange')
plt.xlabel('Wave Numbers')
plt.ylabel('Distance Traveled per Gait Cycle (cm/cycle)')
# plt.title('Rugose with Varied Wave Number')
plt.grid(True)

# Set x-axis ticks using the 'WaveNum' values from the DataFrame
plt.xticks(df['WaveNum'].unique())
# Show the legend with the label for the error bar
plt.legend()
plt.show()
```

```python
plt.rcParams['font.family'] = 'sans-serif'

# Read the data from the Excel file
df = pd.read_excel('plotSpeedData.xlsx')

# Calculate the average and standard deviation of 'Speed (cm/cycle)' for each 'WaveNum'
grouped_df = df.groupby('WaveNum')['Speed (cm/cycle)']
avg_speed = grouped_df.mean()
std_dev = grouped_df.std()

# Plot the first line chart with error bars using the default Matplotlib style
plt.errorbar(avg_speed.index, avg_speed, yerr=std_dev, fmt='o-', label='Flat Surface')

# Read the data from the second Excel file
df_rug = pd.read_excel('plotSpeedDataRug.xlsx')

# Calculate the average and standard deviation of 'Speed (cm/cycle)' for each 'WaveNum' for the second dataset
grouped_df_rug = df_rug.groupby('WaveNum')['Speed (cm/cycle)']
avg_speed_rug = grouped_df_rug.mean()
std_dev_rug = grouped_df_rug.std()

# Plot the second line chart with error bars using the default Matplotlib style
plt.errorbar(avg_speed_rug.index, avg_speed_rug, yerr=std_dev_rug, fmt='o-', label='White Rugose')

plt.xlabel('Wave Numbers')
plt.ylabel('Distance Traveled per Gait Cycle (cm/cycle)')
# plt.title('Flat Surface with Varied Wave Number')
plt.grid(True)

# Set x-axis ticks using the 'WaveNum' values from the DataFrame of the first dataset
plt.xticks(df['WaveNum'].unique())

# Show the legend to differentiate between the two datasets
plt.legend()

plt.show()
```
